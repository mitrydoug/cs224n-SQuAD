```python
import collections
import json

from matplotlib import pyplot as plt
import numpy as np
```

```python
# assumes jupyter launched from notebooks/ directory
with open('../data/train.context', 'r') as f:
    context_lengths = np.array(
        [len(line.strip().split(' ')) for line in f])
```

```python
# assumes jupyter launched from notebooks/ directory
with open('../data/train.question', 'r') as f:
    question_lengths = np.array(
        [len(line.strip().split(' ')) for line in f])
```

```python
with open('../data/train.span', 'r') as f:
    answer_starts = np.array(
        [int(line.strip().split()[0]) for line in f])
with open('../data/train.span', 'r') as f:
    answer_ends = np.array(
        [int(line.strip().split()[1]) for line in f])
with open('../data/train.span', 'r') as f:
    answer_lengths = np.array(
        list(map((lambda t: int(t[1])-int(t[0])+1),
                 (line.strip().split() for line in f))))
```

```python
plt.clf();
plt.figure().set_size_inches(15,10);
plt.title('Answer Length Cumulative Distribution');
plt.hist(answer_lengths, normed=True, cumulative=True, bins=100);
plt.yticks(np.arange(0., 1.01, 0.02));
plt.xticks(np.arange(1, 50, 3))
plt.grid();
```

```python
np.max(answer_lengths)
```

```python
plt.clf();
plt.figure().set_size_inches(15,10);
plt.title('Context Length Cumulative Distribution');
plt.hist(context_lengths, normed=True, cumulative=True, bins=1000);
plt.yticks(np.arange(0., 1.01, 0.02));
plt.xticks(np.arange(0, 800, 20))
plt.grid();
```

```python
plt.clf();
plt.figure().set_size_inches(15,10);
plt.title('Queestion Length Cumulative Distribution');
plt.hist(question_lengths, normed=True, cumulative=True, bins=500);
plt.yticks(np.arange(0., 1.01, 0.02));
plt.xticks(np.arange(0, 60, 2))
plt.grid();
```
